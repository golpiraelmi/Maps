```python
import geopandas as gpd
import matplotlib.pyplot as plt
import geoplot
```

```python
df = gpd.read_file('/Users/golpira/Desktop/Surficial_Geology_of_Alberta,_Generalized_Digital_Mosaic_(DIG_2013-0002).geojson')
```

```python
df.head(2)
```

```python
# Unique entries in GENETIC_GR column
GENETIC_GR = list(df['GENETIC_GR'].unique()) 
GENETIC_GR
```

## Mapping

```python
ax1 = geoplot.choropleth(
    df, projection=geoplot.crs.Orthographic(), 
    hue='GENETIC_GR', cmap='PRGn', figsize=(20, 10), legend=True, 
    legend_kwargs={'bbox_to_anchor': (1.6, 0.5), 'frameon': False})

ax1.set_ylabel('Latitude', fontsize=14)
ax1.set_xlabel('Longitude', fontsize=14)
ax1.set_title ('Surficial Geology of Alberta', fontsize = 14, 
               fontweight ='bold')

# North Arrow
x, y, arrow_length = 0.2, 0.2, 0.06
ax1.annotate('N', xy=(x, y), xytext=(x, y-arrow_length),
            arrowprops=dict(facecolor='black', width=5, headwidth=15),
            ha='center', va='center', fontsize=25,
            xycoords=ax1.transAxes)
plt.savefig('Surficial Geology of Alberta', dpi=600)

plt.show()
```
